```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# MATLAB
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fate($)")
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
# Change the figure size
```

```python
# Object-oriented approach

ax = plt.axes()
ax.plot(x_axis,y_axis,color='g',marker='D',linewidth=2,label="Boston")
ax.set_ylim(0,50)
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.legend()
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis,y_axis, color="green", label= "Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title 
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create a horizontal chart
plt.barh(x_axis,y_axis,color="magenta",label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Month")
plt.title("PyBer Fare per Month")
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt() or the object-oriented approach
fig,ax=plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
# Create horizontal bar-chart using object-oriented
fig,ax=plt.subplots()
# To invert the y-axis, switch the argument
ax.barh(x_axis,y_axis,color='c',label="Chicago")
ax.invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Month")
ax.set_title("PyBer fares per month")
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis,'o')
```

```python
plt.scatter(y_axis,x_axis,color='red',label="Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Fares($)")
plt.ylabel("Months")
plt.title("PyBer Fares per Month")
plt.legend()
plt.xlim(0,50)
```

```python
plt.scatter(x_axis,y_axis,s=y_axis)
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
# Using list comprehension will yield the same result
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
# Object-oriented approach
fig,ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# Bubble charts with the object-oriented approach
fig,ax=plt.subplots()
ax.scatter(y_axis, x_axis, facecolor="blue", edgecolor = "black", alpha = 0.2,label = "Boston", linewidth=2, s=[i*5 for i in y_axis])
ax.invert_yaxis()
ax.set_xlim(0,50)
ax.set_xlabel("Fare($)")
ax.set_ylabel("Month")
ax.set_title("PyBer fares per month")
ax.legend()
```

```python
plt.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')

plt.show()
```
